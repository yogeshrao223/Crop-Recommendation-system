```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "Crop recommendation dataset.csv"   # Update path if needed
df = pd.read_csv(file_path)

# Set Seaborn style
sns.set(style="whitegrid", palette="muted", font_scale=1.1)

# 1. Distribution of Crop Duration
plt.figure(figsize=(8,5))
sns.histplot(df["CROPDURATION"], bins=30, kde=True, color="green")
plt.title("Distribution of Crop Duration")
plt.xlabel("Crop Duration (days)")
plt.ylabel("Frequency")
plt.show()

# 2. Average Soil pH by Crop Type
plt.figure(figsize=(10,6))
sns.boxplot(x="TYPE_OF_CROP", y="SOIL_PH", data=df)
plt.title("Soil pH Variation by Crop Type")
plt.xticks(rotation=45)
plt.show()

# 3. Average Water Requirement per Crop
avg_water = df.groupby("CROPS")["WATERREQUIRED"].mean().sort_values(ascending=False)[:15]
plt.figure(figsize=(12,6))
sns.barplot(x=avg_water.values, y=avg_water.index, palette="viridis")
plt.title("Top 15 Crops by Average Water Requirement")
plt.xlabel("Average Water Required")
plt.ylabel("Crop")
plt.show()

# 4. Temperature vs Humidity Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(x="TEMP", y="RELATIVE_HUMIDITY", hue="SEASON", data=df, alpha=0.6)
plt.title("Temperature vs Humidity by Season")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Relative Humidity (%)")
plt.legend(title="Season")
plt.show()

# 5. Correlation Heatmap of Numerical Features
plt.figure(figsize=(12,8))
corr = df.select_dtypes(include=["float64","int64"]).corr()
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
